## Flight Price Prediction (EDA + Feature engineering)

```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("E:\\flight_price_train_dataset.xlsx")
train_df.head()
```

```python
train_df.shape
```

```python
test_df = pd.read_excel("E:\\flight_price_test_dataset.xlsx")
test_df.head()
```

```python
test_df.shape
```

```python
final_df = pd.concat([train_df,test_df])
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
#feature engineering process 
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
# final_df['Year']=final_df['Date_of_Journey'].apply(lambda x:x.split('/').str[2])
```

```python
final_df.head()
```

```python
final_df['Date']=final_df['Date'].astype('int')
final_df['Month']=final_df['Month'].astype('int')
final_df['Year']=final_df['Year'].astype('int')
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
#remove the date from arrival time bcs their is no use of it
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
```

```python
final_df['Arrival_minute'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype('int')
final_df['Arrival_minute'] = final_df['Arrival_hour'].astype('int')
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
#Now for dep_time 
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute'] = final_df['Dep_Time'].str.split(':').str[1]
final_df['Dep_hour'] = final_df['Dep_hour'].astype('int')
final_df['Dep_minute'] = final_df['Dep_hour'].astype('int')
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].fillna(final_df['Total_Stops'].mode()[0])
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df.head()
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Duration']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
final_df['Airline'] = le.fit_transform(final_df['Airline'])
final_df['Source'] = le.fit_transform(final_df['Source'])
final_df['Destination'] = le.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = le.fit_transform(final_df['Additional_Info'])
```
